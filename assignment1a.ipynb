```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
a=pd.read_csv('C:\\Users\\Suganthi Priya\\Documents\\GitHub\\numpy&pandas\\assignment1.csv')
df=pd.DataFrame(a)
```

```python
x=df.drop(['Price'],axis=1)
y=df['Price']
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
from sklearn.cluster import KMeans
#initialize the object
model=KMeans(n_clusters=2)
#fit the model
model.fit(a)
```

```python
cluster_label=model.labels_
cluster_label
```

```python
plt.scatter(x[np.argwhere(cluster_label==1)],y[np.where(cluster_label==1)],color='pink')
plt.scatter(x[np.argwhere(cluster_label==0)],y[np.where(cluster_label==0)])
```

```python
error=[]
for k in range(1,11):
    model=KMeans(n_clusters=k)
    model.fit(a)
    error.append(model.inertia_)        #total errror  and distortion is avg error=inertia/k
```

```python
plt.scatter(range(1,11),error,color='red')
plt.plot(range(1,11),error)
```

```python
model=KMeans(n_clusters=3)
model.fit(a)
cluster_label=model.labels_
cluster_label
```

```python
plt.scatter(x[np.argwhere(cluster_label==1)],y[np.where(cluster_label==1)],color='red')
plt.scatter(x[np.argwhere(cluster_label==0)],y[np.where(cluster_label==0)])
plt.scatter(x[np.argwhere(cluster_label==2)],y[np.where(cluster_label==2)],color='green')
```
